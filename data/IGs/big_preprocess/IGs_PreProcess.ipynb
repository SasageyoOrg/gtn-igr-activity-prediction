```python
def parse_graphs(input_data):

  graphs = []
  current_graph = {'nodes': [], 'edges': []}

  for line in input_data.split("\n"):
      line = line.strip()
      if line.strip():
        if line.startswith("XP"):
            if current_graph['nodes'] or current_graph['edges']:
                graphs.append(current_graph)
                current_graph = {'nodes': [], 'edges': []}
        elif line:
            elements = line.split()

            if elements[0] == "v":
                node_id = int(elements[1])
                node_label = elements[2]
                current_graph['nodes'].append((node_id, node_label))
            elif elements[0] == "e":
                source_id = int(elements[1])
                target_id = int(elements[2])
                edge_label = elements[3]
                current_graph['edges'].append((source_id, target_id, edge_label))

  if current_graph['nodes'] or current_graph['edges']:
      graphs.append(current_graph)

  return graphs


def build_pairs_from_graphs_dfs(graphs):
    pairs = []
    pair_set = set()

    for graph in graphs:
        nodes = graph['nodes']
        edges = graph['edges']

        # Find the starting and ending nodes
        start_node = next(node for node in nodes if node[1] == 'start')
        end_node = next(node for node in nodes if node[1] == 'end')

        # Perform DFS to explore all paths
        stack = [(start_node, [])]

        while stack:
            node, path = stack.pop()

            path.append(node[1])
            if len(path) >= 3:
              if path[-3:] == [path[-1]] * 3:
                  path.pop(-1)

            if len(path) < 2:
                stack.append((nodes[1], path.copy()))
                continue


            # Create a pair for the current prefix and next activity
            # prefix = path.copy()
            prefix = tuple(path)

            # Find the outgoing edges from the current node
            outgoing_edges = [edge for edge in edges if edge[0] == node[0]]
            for edge in outgoing_edges:
                target_node = next(n for n in nodes if n[0] == edge[1])
                next_activity = target_node[1]
                if not (prefix[-1] == prefix[-2] == next_activity):
                    pair = (prefix, next_activity)
                    if pair not in pair_set:
                      pairs.append(pair)
                      pair_set.add(pair)

                # Check if we reached the end node
                if target_node == end_node:
                    continue

                stack.append((target_node, path.copy()))
    return pairs



def convert_graph_to_format(graph_pair):
    graph, label = graph_pair
    num_nodes = len(graph)
    result = f"{num_nodes} {label}"

    for i, node in enumerate(graph):
        neighbors = []
        if i > 0:
            neighbors.append(str(i-1))   # was graph[i-/+1]
        if i < num_nodes - 1:
            neighbors.append(str(i+1))

        num_neighbors = len(neighbors)
        neighbor_list = ' '.join(neighbors)
        result += f"\n{node} {num_neighbors} {neighbor_list}"

    return result

def create_vocabulary(pairs):
    vocabulary = {}
    index = 0

    for graph, label in pairs:
        for node in graph:
            if node not in vocabulary:
                vocabulary[node] = index
                index += 1
        if label not in vocabulary:
            vocabulary[label] = index
            index += 1

    return vocabulary

def translate_graphs_with_conversion(pairs, vocabulary):
    translated_pairs = []

    for graph, label in pairs:
        translated_graph = [vocabulary[node] for node in graph]
        translated_label = vocabulary[label] if label in vocabulary else label

        num_nodes = len(translated_graph)
        formatted_graph = f"{num_nodes} {translated_label}"

        for i, node in enumerate(translated_graph):
            neighbors = []
            if i > 0:
                neighbors.append(str(i-1)) # was translated_graph[i-/+1]
            if i < num_nodes - 1:
                neighbors.append(str(i+1))

            num_neighbors = len(neighbors)
            formatted_neighbors = " ".join(map(str, neighbors))
            formatted_node = f"{node} {num_neighbors} {formatted_neighbors}"
            formatted_graph += "\n" + formatted_node

        translated_pairs.append(formatted_graph)

    return translated_pairs


def translate_graphs(pairs, vocabulary):

    translated_pairs = []

    for graph, label in pairs:
        translated_graph = [vocabulary[node] for node in graph]
        translated_label = vocabulary[label] if label in vocabulary else label

        translated_pairs.append(f"{translated_graph} {translated_label}")
        #translated_pairs.append(tuple(translated_graph, [translated_label]))

    return translated_pairs


def write_graphs_to_file(translated_graphs, filename):
    with open(filename, 'w') as file:
        #file.write(str(len(translated_graphs)) + '\n')
        for graph in translated_graphs:
            file.write(graph + '\n')

def limit_self_loops(pairs):
    result_prefixes = []
    result_label = pairs[1]
    prefix = pairs[0]

    for i in range(len(prefix)):
        if i >= 2 and prefix[i] == prefix[i-1] and prefix[i] == prefix[i-2]:
            continue
        result_prefixes.append(prefix[i])

    return result_prefixes, result_label


def filter_pairs(pairs):
    filtered_pairs = []
    for pair in pairs:
        pair = pair.strip().split(' ')
        prefix = eval(pair[0])
        label = int(pair[1])

        if len(prefix) > 1 and prefix[-1] == prefix[-2] == label:
            prefix = [x for i, x in enumerate(prefix) if i == len(prefix) - 1 or prefix[i] != prefix[i + 1]]

        filtered_pairs.append((prefix, label))

    return filtered_pairs
```

```python
vocabulary = create_vocabulary(pairs)
translated_pairs = translate_graphs(pairs, vocabulary)
print(f"Vocabulary: {vocabulary}")
```
